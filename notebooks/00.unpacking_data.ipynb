```python
# Import packages
import zipfile
import os
import pandas as pd
import numpy as np
```

```python
# List of ZIP file paths
zip_files = [
    '/content/drive/MyDrive/case_rentcars/data/raw/aisles.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/departments.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/order_products__prior.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/order_products__train.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/orders.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/products.csv.zip',
    '/content/drive/MyDrive/case_rentcars/data/raw/sample_submission.csv.zip'
]

# Directory to extract files to
extract_to_directory = '/content/drive/MyDrive/case_rentcars/data/interim/'

# Ensure the target directory exists
os.makedirs(extract_to_directory, exist_ok=True)

# Extract each ZIP file
for zip_file_path in zip_files:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to_directory)

print(f'Files extracted to {extract_to_directory}')
```
